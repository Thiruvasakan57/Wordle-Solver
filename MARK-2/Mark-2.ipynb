# Wordle Solver

```python
import pandas as pd
import numpy as np
```

```python
#db is going to be the working database of wordle words
db = pd.read_csv('words.txt', sep =' ', header = None, names = ['Words'])
db
```

```python
df = db['Words'].str.split('',expand=True)
df = df.drop(columns = [0,6])
df['Word'] = db.copy()
```

```python
for zeta in range(0,7):
    print('TRIAL: '+str(zeta+1)+ '\n')
    guess = input("Guess Word: \n")
    clue = input("Enter Clue: \n")
    for i in range(0, len(clue)) :
        if(clue[i]=='b'):
            for j in range(1, df.shape[1]):
                indexes = df[df[j].str.contains(guess[i])==True].index
                df.drop(indexes, inplace = True)
                df.to_csv("Trial "+str(zeta+1)+' Case B'+'.csv')
                
        if(clue[i]=='g'):
                indexes = df[df[i+1].str.contains(guess[i])==False].index
                df.drop(indexes, inplace = True)
                df.to_csv("Trial "+str(zeta+1)+' Case G'+'.csv')
                
        if(clue[i]=='y'):
            indexes = df[df[i+1].str.contains(guess[i]) == True].index
            df.drop(indexes, inplace  = True)
            df.to_csv("Trial "+str(zeta+1)+' Case Y1'+'.csv')
            indexes = df[df['Word'].str.contains(guess[i])==False].index
            df.to_csv("Trial "+str(zeta+1)+' Case Y2'+'.csv')
            df.drop(indexes, inplace = True)
    print("Possible Words are: \n", df, "\n"*5)
    
```
