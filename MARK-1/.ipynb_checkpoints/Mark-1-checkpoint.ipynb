# Wordle Solver

```python
import pandas as pd
```

```python
#db is going to be the working database of wordle words
db = pd.read_csv('words.txt', sep =' ', header = None, names = ['Words'])
db
```

```python
df = db['Words'].str.split('',expand=True)
df = df.drop(columns = [0,6])
df
```

```python
for zeta in range(0,7):
    print('TRIAL: '+str(zeta+1)+ '\n')
    zeta = zeta + 1
    dnc = input('Input non existant characters: ')
    cnc = input('Input existing characters: ')
    for i in range(0,len(dnc)):
        indexes = db[db['Words'].str.contains(dnc[i])==True].index
        db.drop(indexes, inplace = True)
    for j in range(0,len(cnc)):
        indexes = db[db['Words'].str.contains(cnc[j])==False].index
        db.drop(indexes, inplace = True)
        
    print('Possible words are: \n')
    print(db)
    df = db.copy()
    df = db['Words'].str.split('', expand = True)
    df.drop(columns = [0,6])
    df.to_csv("Trial_"+str(zeta-1)+".csv")
    print('\n'*5)
```

R - not in 1
O - not in 2
A - not in 3,4
